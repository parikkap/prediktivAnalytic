```python
def calculateMedian(values):
    
    if(type(values) != String)
    local_copy = list(values)
    local_copy.sort()
    center   = len(local_copy)/2
    
    if len(local_copy) % 2:
        return float(local_copy[int(center)])

    return (local_copy[int(center-1.0)]+local_copy[int(center)])/2.0
```

```python
result = calculateMedian(sensor_df.get('temp_min'))
result
```

```python
def getMedianOfDataframeHeaders(dataframe):
    headers = list(dataframe.columns.values)
    for header in headers:
        result = calculateMedian(header)
        print(result)
```

```python
getMedianOfDataframeHeaders(sensor_df)
```
